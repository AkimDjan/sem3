### 1. Расчет постоянной времени $\tau$

В схеме стоит интегрирующая цепочка $R_2 C$. Она нужна, чтобы превратить сигнал, пропорциональный ЭДС (производной потока), в сигнал, пропорциональный самому магнитному потоку $B$.

**Дано:**
*   $R_2 = 22\,\text{кОм} = 22\,000\,\text{Ом}$
*   $C = 4\,\text{мкФ} = 4 \cdot 10^{-6}\,\text{Ф}$

**Формула:**
$$ \tau = R_2 \cdot C $$

**Расчет:**
$$ \tau = 22\,000 \cdot 4 \cdot 10^{-6} = 88 \cdot 10^{-3}\,\text{с} = \mathbf{0.088\,\text{с}} $$
*(Или 88 миллисекунд).*

---

### 2. Проверка условия работы интегратора

*«Убедитесь в выполнении неравенства $R_2 \gg \frac{1}{\omega C}$»*.
Это условие того, что цепочка действительно работает как интегратор, а не как делитель напряжения.

*   $\omega = 2\pi f$ — циклическая частота.
$f = 500\,\text{Гц}$ (это «худший» случай, так как сопротивление конденсатора $\frac{1}{\omega C}$ максимально на низкой частоте).

**Расчет емкостного сопротивления:**
$$ X_c = \frac{1}{2\pi \cdot 500 \cdot 4 \cdot 10^{-6}} \approx \frac{1}{3140 \cdot 4 \cdot 10^{-6}} \approx \frac{1}{0.01256} \approx \mathbf{79.6\,\text{Ом}} $$

**Сравнение:**
$$ R_2 = 22\,000\,\text{Ом} $$
$$ X_c \approx 80\,\text{Ом} $$

**Вывод:** $22\,000 \gg 80$. Условие выполняется с огромным запасом (в 275 раз). Значит, схема работает корректно, и мы имеем право использовать формулу для $B$.

---

### 3. Расчет зависимости $H = \phi(U_1)$

$U_1$ (или $U_x$ на схеме) — это напряжение на резисторе $R_1$, которое подается на горизонтальный вход осциллографа. Оно пропорционально току: $U_1 = I_1 \cdot R_1$, откуда $I_1 = \frac{U_1}{R_1}$.

Закон полного тока для тороида:
$$ H \cdot l = W_1 \cdot I_1 $$
где $l$ — длина средней линии тороида.
$$ l = \pi \cdot d = 3.14159 \cdot 0.025\,\text{м} \approx \mathbf{0.0785\,\text{м}} $$

Подставляем ток $I_1$:
$$ H = \frac{W_1}{l} \cdot \frac{U_1}{R_1} $$

Теперь подставим числа, чтобы получить простую формулу вида $H = K \cdot U_1$:
$$ H = \frac{150}{0.0785 \cdot 11\,000} \cdot U_1 $$
$$ H = \frac{150}{863.5} \cdot U_1 $$
$$ H \approx \mathbf{0.174 \cdot U_1} $$

---




$$ B = \frac{R_2 \cdot C}{W_2 \cdot S} \cdot U_y $$
Подставляем константы ($R_2=22\text{к}, C=4\text{мкФ}, W_2=300, S=9\text{мм}^2$):
$$ B = \frac{22000 \cdot 4 \cdot 10^{-6}}{300 \cdot 9 \cdot 10^{-6}} \cdot U_y = \frac{0.088}{0.0027} \cdot U_y \approx \mathbf{32.6 \cdot U_y \, [\text{Тл}]} $$

1.  Чтобы получить **A/м**, умножаем вольты с оси X на **0.174**.
2.  Чтобы получить **Тл**, умножаем вольты с оси Y на **32.6**.

## 2. Расчет потерь на гистерезис

### 2.1. Методика расчета
Нам нужно найти удельные потери энергии $q$ (Джоуль на кубический метр) за один цикл перемагничивания.


**1. Определение площади петли в «Вольтах» ($S_{volt}$)**
Мы считаем количество клеток ($N$), которое занимает петля на экране, и умножаем на цену деления по осям X и Y:
$$ S_{volt} = N \cdot (\text{Масштаб}_X) \cdot (\text{Масштаб}_Y) \quad [\text{В}^2] $$

**2. Перевод в физические величины ($q$)**
Мы знаем калибровочные коэффициенты прибора (рассчитанные в п.1):
*   $K_H = 0.174$ А/(м·В) — коэффициент по полю.
*   $K_B = 32.59$ Тл/В — коэффициент по индукции.

Перемножив их, получаем **общий коэффициент перевода ($K_{total}$)**:
$$ K_{total} = 0.174 \cdot 32.59 \approx \mathbf{5.67} \, \frac{\text{Дж}}{\text{м}^3 \cdot \text{В}^2} $$

Теперь просто умножаем площадь в вольтах на этот коэффициент:
$$ q = S_{volt} \cdot 5.67 \quad [\text{Дж}/\text{м}^3] $$

**3. Расчет полной энергии ($Q$)**
Удельные потери умножаем на объем ферромагнетика ($V \approx 7.07 \cdot 10^{-7} \text{м}^3$):
$$ Q = q \cdot V \cdot 10^6 \quad [\text{мкДж}] $$

```python
import pandas as pd
import numpy as np

# 1. Константы и коэффициенты
scale_x = 10.0      # В/дел
scale_y = 0.005     # В/дел (5 мВ)
cell_area_volt = scale_x * scale_y  # 0.05 В^2
K_total = 5.67      # Коэффициент перевода (Дж/м^3) / В^2
Volume = 7.07e-7    # Объем, м^3

# 2. Данные (N) с картинки
data = {
    'f_Hz': [100, 200, 300, 500, 700, 900, 1100],
    'N_cells': [23.0, 27.0, 28.5, 24.0, 17.5, 13.5, 11.0]
}

# 3. Создаем DataFrame
df = pd.DataFrame(data)

# 4. Расчеты
# Площадь в Вольтах = N * 0.05
df['S_volt (V^2)'] = df['N_cells'] * cell_area_volt

# Удельные потери q = S_volt * 5.67
df['q (J/m3)'] = df['S_volt (V^2)'] * K_total

# Полная энергия Q = q * Volume * 10^6 (в мкДж)
df['Q_total (uJ)'] = df['q (J/m3)'] * Volume * 1e6

# 5. Округление для красивого вывода
df_display = df.copy()
df_display['S_volt (V^2)'] = df_display['S_volt (V^2)'].round(3)
df_display['q (J/m3)'] = df_display['q (J/m3)'].round(2)
df_display['Q_total (uJ)'] = df_display['Q_total (uJ)'].round(2)

df_display
```

```python
import pandas as pd
import numpy as np

# 1. Константы
K_total = 5.67      # Коэффициент перевода (Дж/м^3) / В^2
Volume = 7.07e-7    # Объем, м^3

# 2. Данные для 7 файлов (28-34)
# ВАЖНО: Масштабы (Scale_X, Scale_Y) менялись!
data = {
    # Напряжение амплитуды (Ux)
    'Ux_V': [6.0, 10.0, 14.0, 20.0, 25.0, 30.0, 38.0],

    # Масштаб по X (В/дел)
    'Scale_X': [2.0, 5.0, 5.0, 10.0, 10.0, 10.0, 10.0],

    # Масштаб по Y (В/дел). Внимание: 1mV = 0.001 В
    'Scale_Y': [0.001, 0.001, 0.002, 0.005, 0.005, 0.005, 0.005],

    # Оценка площади в клетках (Можешь поправить эти числа!)
    'N_cells': [4.5, 8.5, 16.5, 12.5, 19.0, 23.0, 27.0]
}

# 3. Создаем DataFrame
df = pd.DataFrame(data)

# 4. Расчеты

# Считаем цену одной клетки для каждой строки отдельно (Scale_X * Scale_Y)
df['Cell_Val (V^2)'] = df['Scale_X'] * df['Scale_Y']

# Площадь петли в Вольтах = N * Цена_Клетки
df['S_volt (V^2)'] = df['N_cells'] * df['Cell_Val (V^2)']

# Удельные потери q = S_volt * 5.67
df['q (J/m3)'] = df['S_volt (V^2)'] * K_total

# Полная энергия Q = q * Volume * 10^6
df['Q_total (uJ)'] = df['q (J/m3)'] * Volume * 1e6

# 5. Округление и вывод
df_disp = df[['Ux_V', 'N_cells', 'Cell_Val (V^2)', 'S_volt (V^2)', 'q (J/m3)', 'Q_total (uJ)']].copy()

df_disp['S_volt (V^2)'] = df_disp['S_volt (V^2)'].round(4)
df_disp['q (J/m3)'] = df_disp['q (J/m3)'].round(3)
df_disp['Q_total (uJ)'] = df_disp['Q_total (uJ)'].round(3)

df_disp
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

f_Hz = np.array([100, 200, 300, 500, 700, 900, 1100])
N_cells = np.array([23.0, 27.0, 28.5, 24.0, 17.5, 13.5, 11.0])

# === 2. РАСЧЕТЫ ===
# Множитель площади: 10 В * 0.005 В * 5.67 = 0.2835
scale_factor = 0.2835
q_vals = N_cells * scale_factor

# === 3. АППРОКСИМАЦИЯ (Только линейный режим) ===
# Берем данные только от 500 Гц и выше
mask = f_Hz >= 500
f_fit = f_Hz[mask]
q_fit = q_vals[mask]

def power_law(x, a, n):
    return a * np.power(x, n)

# Фитируем только по хвосту графика
popt, _ = curve_fit(power_law, f_fit, q_fit)

# === 4. ПОСТРОЕНИЕ ГРАФИКА ===
plt.figure(figsize=(10, 7))

# 1. Рисуем все экспериментальные точки
plt.plot(f_Hz, q_vals, 'bo', markersize=8, label='Эксперимент')


# 2. Рисуем аппроксимацию (продлим её немного назад до 400, чтобы было видно начало)
f_smooth = np.linspace(450, 1150, 100)
plt.plot(f_smooth, power_law(f_smooth, *popt), 'r--', linewidth=2,
         label=f'Аппроксимация (500-1100 Гц):\n$q \\approx {popt[0]:.1f} \\cdot f^{{{popt[1]:.2f}}}$')

# Оформление
plt.title("Зависимость удельных потерь $q$ от частоты $f$", fontsize=14)
plt.xlabel("Частота $f$, Гц", fontsize=12)
plt.ylabel("Удельные потери $q$, Дж/м³", fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend(fontsize=11)

# Подписи значений
for x, y in zip(f_Hz, q_vals):
    plt.text(x, y + 0.3, f"{y:.2f}", ha='center', fontsize=9)

plt.show()

print(f"Уравнение аппроксимации: q = {popt[0]:.2f} * f^({popt[1]:.3f})")
```



### Теоретическое обоснование степенной зависимости $q(f) = A \cdot f^{-n}$

Наш результат ($n \approx 0.6 \dots 0.9$) говорит о том, что потери падают почти обратно пропорционально частоте. Обоснуем это через **частотную дисперсию магнитной проницаемости**.

#### 1. Связь потерь с индукцией
Площадь петли гистерезиса (удельные потери $q$) для эллиптического цикла можно оценить как:
$$ q = \pi \cdot H_{max} \cdot B_{max} \cdot \sin(\delta) $$
Где:
*   $H_{max}$ — амплитуда поля (в нашем опыте **константа**, т.к. амплитуда генератора постоянна).
*   $B_{max}$ — амплитуда индукции.
*   $\delta$ — угол потерь (фазовый сдвиг).

Поскольку $\sin(\delta)$ меняется слабо (в нашем опыте он около $0.7 \dots 0.9$), основной вклад в изменение потерь вносит поведение амплитуды **индукции $B_{max}$**.

Следовательно, зависимость потерь $q(f)$ определяется зависимостью магнитной проницаемости $\mu(f)$, так как $B_{max} = \mu(f) \cdot H_{max}$.

#### 2. Релаксационная модель (Дебая)
Ферромагнетик не реагирует на поле мгновенно. Процессы смещения доменных границ и вращения магнитных моментов имеют инерцию. Это описывается теорией релаксации.

В простейшем случае (релаксация Дебая) зависимость магнитной проницаемости от частоты выглядит так:
$$ \mu(f) = \frac{\mu_{stat}}{\sqrt{1 + (2\pi f \tau)^2}} $$
Где:
*   $\tau$ — время релаксации.
*   $\mu_{stat}$ — статическая проницаемость (на 0 Гц).

**Рассмотрим два случая:**
*   **Низкие частоты ($2\pi f \tau \ll 1$):** $\mu \approx \text{const}$. Потери не должны зависеть от частоты.
*   **Высокие частоты ($2\pi f \tau \gg 1$):** Единицей в знаменателе можно пренебречь. Тогда:
    $$ \mu(f) \approx \frac{\mu_{stat}}{2\pi \tau \cdot f} \sim \frac{1}{f} $$

**Теоретический вывод:** При частотах, превышающих обратное время релаксации ($f > 1/2\pi\tau$), магнитная проницаемость (а значит и индукция $B$, и потери $q$) должна убывать обратно пропорционально частоте, то есть как $f^{-1}$.

#### 3. Почему у нас показатель степени $n \neq 1$, а дробный ($0.6 \dots 0.9$)?
В реальных поликристаллических материалах (как наш пермаллой) нет **одного** строгого времени релаксации $\tau$.
*   В материале есть домены разных размеров.
*   Есть дефекты кристаллической решетки, которые по-разному тормозят доменные стенки.

Поэтому вместо одной $\tau$ существует **спектр времен релаксации**. В физике твердого тела это описывается эмпирической формулой Коула-Коула, которая приводит к степенной зависимости с дробным показателем:
$$ \mu(f) \sim \frac{1}{f^{1-\alpha}} $$
Где параметр $\alpha$ характеризует "ширину" распределения времен релаксации.

### Итоговое резюме для отчета:

> Экспериментально полученная зависимость вида $q \sim f^{-n}$ (где $0 < n < 1$) обусловлена тем, что мы работаем в частотном диапазоне **магнитной дисперсии**.
>
> Из-за инерционности доменных процессов магнитная проницаемость материала падает с ростом частоты. Согласно релаксационной теории, в идеальном случае это падение описывается законом $1/f$. Наблюдаемое отклонение показателя степени от единицы ($n \approx 0.7$) объясняется наличием в образце широкого спектра времен релаксации (разброс размеров доменов и дефектов структуры), что характерно для реальных ферромагнетиков.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. ДАННЫЕ (7 ТОЧЕК) ===
# Напряжения Ux (В)
Ux_vals = np.array([6.0, 10.0, 14.0, 20.0, 25.0, 30.0, 38.0])
# Твои оценки площади N (клеток)
N_cells = np.array([4.5, 8.5, 16.5, 12.5, 19.0, 23.0, 27.0])
# Масштабы (Цена клетки в В^2)
scale_factors = np.array([0.002, 0.005, 0.010, 0.050, 0.050, 0.050, 0.050])

# === 2. РАСЧЕТЫ ===
K_H = 0.174         # Коэфф. для перевода Вольт в А/м
K_total = 5.67      # Коэфф. для перевода площади в Дж/м3

# Считаем физические величины
# H - это и есть "Амплитуда поля"
H_vals = Ux_vals * K_H
q_vals = N_cells * scale_factors * K_total

# === 3. АППРОКСИМАЦИЯ (Только начальный участок) ===
# Берем первые 3 точки (область слабых полей, закон Рэлея)
H_fit = H_vals[:3]
q_fit = q_vals[:3]

def power_law(x, a, m):
    return a * np.power(x, m)

popt, _ = curve_fit(power_law, H_fit, q_fit)

# === 4. ПОСТРОЕНИЕ ГРАФИКА ===
plt.figure(figsize=(10, 7))

# Экспериментальные точки (все 7)
plt.plot(H_vals, q_vals, 'ro', markersize=9, label='Эксперимент (все амплитуды)')

# Теоретическая кривая (продлеваем, чтобы показать отклонение)
h_smooth = np.linspace(0.5, 7.0, 100)
plt.plot(h_smooth, power_law(h_smooth, *popt), 'k--', linewidth=2,
         label=f'Аппроксимация (по первым 3 точкам):\n$q \\approx {popt[0]:.2f} \\cdot H^{{{popt[1]:.2f}}}$')

# Оформление
plt.title("Зависимость потерь за период $q$ от амплитуды поля $H$", fontsize=14)
plt.xlabel("Амплитуда напряженности поля $H$, А/м", fontsize=12)
plt.ylabel("Удельные потери $q$, Дж/м³", fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend(fontsize=11)

# Подписи значений
for x, y in zip(H_vals, q_vals):
    plt.text(x, y + 0.3, f"{y:.2f}", ha='center', fontsize=9)

plt.show()

print(f"Степенной показатель (по начальному участку): m = {popt[1]:.3f}")
```


### Методика расчета сдвига фаз ($\phi$)

Для расчета фазового сдвига между напряженностью магнитного поля ($H \sim U_x$) и магнитной индукцией ($B \sim U_y$) используется геометрическое свойство фигур Лиссажу.

В линейном режиме гистерезиса петля имеет форму эллипса. Площадь этого эллипса ($S$) связана с амплитудами сигналов и разностью фаз следующим соотношением:

$$ S = \pi \cdot U_x \cdot U_y \cdot \sin(\phi) $$

Отсюда выражаем синус угла сдвига фаз:

$$ \sin(\phi) = \frac{S_{volt}}{\pi \cdot U_x \cdot U_y} $$

**Где:**
*   **$S_{volt}$** — Площадь петли в «Вольт-квадратах» ($В^2$). Определяется экспериментально путем подсчета клеток $N$ и умножения на масштаб осей:
    $$ S_{volt} = N_{\text{клеток}} \cdot (\text{Масштаб}_X \cdot \text{Масштаб}_Y) $$
*   **$U_x$** — Амплитуда напряжения на горизонтальном входе (половина полной ширины петли), В.
*   **$U_y$** — Амплитуда напряжения на вертикальном входе (половина полной высоты петли), В.
*   **$\pi$** — Математическая константа ($\approx 3.14159$).

**Физический смысл:**
*   Если $\phi = 0^\circ$, то $\sin(\phi) = 0$, площадь $S = 0$. Петля вырождается в прямую линию (потерь нет).
*   Если $\phi = 90^\circ$, то $\sin(\phi) = 1$, площадь максимальна ($S = \pi U_x U_y$). Петля становится "максимально широким" эллипсом.
